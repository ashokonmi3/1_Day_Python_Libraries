## üìö Theory: Why labeling and formatting plots matter

- **Labels** help us understand what the axes represent (e.g., time, sales, temperature).
- **Titles** give the plot a name so viewers know what the graph is about.
- **Legends** explain what different colors or lines in the plot mean.
- **Formatting** improves readability and makes plots look neat and professional.

Without clear labels, titles, and legends, graphs can be confusing and misleading! üö´‚ùì

Let's learn how to add these important elements using matplotlib! üõ†Ô∏è

## 1Ô∏è‚É£ Adding Titles and Axis Labels

We use:
- `plt.title()` to add a plot title.
- `plt.xlabel()` to label the x-axis.
- `plt.ylabel()` to label the y-axis.

### Example: Simple line plot with labels

```python
import matplotlib.pyplot as plt

# Sample data
days = [1, 2, 3, 4, 5]
sales = [10, 15, 8, 20, 25]

# Create line plot
plt.plot(days, sales, marker='o', color='blue')  # Line with dots

# Add title and axis labels
plt.title('Sales Over 5 Days')  # Title
plt.xlabel('Day')  # X-axis label
plt.ylabel('Sales (units)')  # Y-axis label

# Show plot
plt.show()
```

**Explanation:**
- Title tells us this graph shows sales over 5 days.
- X-axis labeled "Day" shows what the horizontal numbers mean.
- Y-axis labeled "Sales (units)" tells us what vertical values mean.


## 2Ô∏è‚É£ Adding Legends

- When a plot has multiple lines or data series, legends explain which line is which.
- Use `label='name'` in plotting function and then call `plt.legend()`.

### Example: Two line plot with legend

```python
# Data for two products
days = [1, 2, 3, 4, 5]
sales_product_A = [10, 15, 8, 20, 25]
sales_product_B = [5, 7, 6, 10, 15]

# Plot both lines with labels
plt.plot(days, sales_product_A, marker='o', color='blue', label='Product A')
plt.plot(days, sales_product_B, marker='s', color='red', label='Product B')

# Add title, labels
plt.title('Sales of Products A and B Over 5 Days')
plt.xlabel('Day')
plt.ylabel('Sales (units)')

# Show legend
plt.legend()

# Show plot
plt.show()
```

**Explanation:**
- `label` inside `plt.plot()` names each line.
- `plt.legend()` displays the legend box with these names.
- Markers differ (`o` for circle, `s` for square) for clarity.


## 3Ô∏è‚É£ Formatting plots

We can format plots to make them more readable or visually appealing:
- Change colors
- Change line styles (`-`, `--`, `:`)
- Change marker styles
- Add grid lines
- Set axis limits

### Example: Customized line plot

```python
# Data
x = [1, 2, 3, 4, 5]
y = [10, 12, 9, 14, 15]

# Plot with formatting
plt.plot(x, y, color='purple', linestyle='--', marker='^', markersize=10, label='Trend')

# Add labels and title
plt.title('Customized Plot Example')
plt.xlabel('X axis')
plt.ylabel('Y axis')

# Add grid
plt.grid(True)

# Set limits on axes
plt.xlim(0, 6)
plt.ylim(5, 20)

# Show legend
plt.legend()

# Show plot
plt.show()
```

**Explanation:**
- Color is purple.
- Line is dashed (`--`).
- Markers are triangles (`^`) sized 10.
- Grid lines added for easy reading.
- Axis limits set to control visible area.
